```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
from pandas.core.dtypes.common import is_numeric_dtype
```

```python
for col in df1.columns:
    if is_numeric_dtype(df1[col]):
        continue
    else:
        df1[col] = le.fit_transform(df1[col])
```

```python
df1.tail()
```

**Training and Testing Sets**

```python
from sklearn.model_selection import train_test_split
```

```python
X = df1.drop('DRK_YN', axis=1)
Y = df1['DRK_YN']
```

```python
Xtrain , Xtest, ytrain, ytest = train_test_split(X, Y ,test_size=0.3, random_state=42)
```

```python
Xtrain.shape
```

```python
Xtest.shape
```

K-Nearest Neighbors (KNN)

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=833)
```

```python
knn.fit(Xtrain, ytrain)
```

```python
knn_predictions = knn.predict(Xtest)
```

 Evaluate the Models

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, classification_report

# Evaluate the KNN model
knn_accuracy = accuracy_score(ytest, knn_predictions)
knn_precision_score = precision_score(ytest, knn_predictions)
knn_recall_score = recall_score(ytest, knn_predictions)
knn_cm = confusion_matrix(ytest, knn_predictions)
knn_classification_report = classification_report(ytest, knn_predictions)
```

```python
print("K-Nearest Neighbors Accuracy score:")
print(knn_accuracy)
print("\nK-Nearest Neighbors Precision score:")
print(knn_precision_score)
print("\nK-Nearest Neighbors Recall score:")
print(knn_recall_score)
print("\nK-Nearest Neighbors Confusion Matrix:")
print(knn_cm)
print("\nK-Nearest Neighbors Classification Report:")
print(knn_classification_report)
```

**Decision Tree Classifier**

```python
from sklearn.tree import DecisionTreeClassifier

# Create and train the Decision Tree classifier
decision_tree = DecisionTreeClassifier(random_state=42)
decision_tree.fit(Xtrain, ytrain)
```

```python
# Make predictions on the testing data
dt_predictions = decision_tree.predict(Xtest)
```

```python
# Evaluate the Decision Tree model
dt_accuracy = accuracy_score(ytest, dt_predictions)
dt_precision_score = precision_score(ytest, dt_predictions)
dt_recall_score = recall_score(ytest, dt_predictions)
dt_cm = confusion_matrix(ytest, dt_predictions)
dt_classification_report = classification_report(ytest, dt_predictions)
```

```python
print("Decision Tree Accuracy score:")
print(dt_accuracy)
print("\nDecision Tree Precision score:")
print(dt_precision_score)
print("\nDecision Tree Recall score:")
print(dt_recall_score)
print("\nDecision Tree Confusion Matrix:")
print(dt_cm)
print("\nDecision Tree Classification Report:")
print(dt_classification_report)
```

```python
print("K-Nearest Neighbors Confusion Matrix:")
print(knn_cm)
print("K-Nearest Neighbors Classification Report:")
print(knn_classification_report)

print("\nDecision Tree Confusion Matrix:")
print(dt_cm)
print("Decision Tree Classification Report:")
print(dt_classification_report)
```
