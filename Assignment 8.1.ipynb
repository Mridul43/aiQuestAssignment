```python
import numpy as np
import pandas as pd
from pandas.core.dtypes.common import is_numeric_dtype
```

```python
df = pd.read_csv('insurance.csv')
df.head()
```

```python
df1 = df.copy()
df2 = df.copy()
df3 = df.copy()
df4 = df.copy()
```

# Label Encoder

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
for col in df1.columns:
    if is_numeric_dtype(df1[col]):
        continue
    else:
        df1[col] = le.fit_transform(df1[col])
```

```python
df1.head()
```

# One Hot

```python
df2.head()
```

```python
for col in df2.columns:
    if is_numeric_dtype(df2[col]):
        continue
    else:
        one = pd.get_dummies(df2[col], prefix=col)
        df2.drop(df2[[col]], axis=1, inplace=True)
        df2 = pd.concat([df2, one], axis=1)
```

```python
df2.head()
```

# Ordinal

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
enc = OrdinalEncoder()
```

```python
for col in df3.columns:
    if is_numeric_dtype(df3[col]):
        continue
    else:
        df3[col] = enc.fit_transform(df3[col])
```
